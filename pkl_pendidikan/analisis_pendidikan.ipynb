```python
import pandas as pd
import matplotlib.pyplot as plt

# ============================
# LOAD DATA
# ============================
df = pd.read_csv("pendidikan_2024.csv")

bulan_list = [
    "Januari","Februari","Maret","April","Mei","Juni",
    "Juli","Agustus","September","Oktober","November","Desember"
]

for bulan in bulan_list:
    data_bulan = df[df['bulan'] == bulan]
    print(f"\n========== BULAN {bulan.upper()} ==========")

    # hitung data
    pendidikan_counts = data_bulan['tingkat_pendidikan'].value_counts().to_dict()
    bantuan_counts    = data_bulan['jenis_bantuan'].value_counts().to_dict()

    # ============================
    # PIE CHART KIRI & KANAN
    # ============================
    plt.figure(figsize=(12,6))

    # kiri
    plt.subplot(1,2,1)
    plt.pie(pendidikan_counts.values(),
            labels=pendidikan_counts.keys(),
            autopct='%1.1f%%')
    plt.title("Tingkat Pendidikan")

    # kanan
    plt.subplot(1,2,2)
    plt.pie(bantuan_counts.values(),
            labels=bantuan_counts.keys(),
            autopct='%1.1f%%')
    plt.title("Jenis Bantuan")

    plt.suptitle(f"Distribusi Data - {bulan}")
    plt.show()

    # ============================
    # TABEL OUTPUT BERSIH
    # ============================
    print("Jumlah Tingkat Pendidikan:")
    for k, v in pendidikan_counts.items():
        print(f"- {k} : {v}")

    print("\nJumlah Jenis Bantuan:")
    for k, v in bantuan_counts.items():
        print(f"- {k} : {v}")

    # ============================
    # TOTAL
    # ============================
    total = len(data_bulan)
    print(f"\nTotal Mustahik Bulan {bulan}: {total}")
```
