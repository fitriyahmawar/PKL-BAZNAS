```python
import pandas as pd
from IPython.display import display

# Load dataset
df = pd.read_csv("mustahik_kelengkapan.csv")

# Aturan keputusan
def hasil_keputusan(row):
    if row['ktp'] == 1 and row['kk'] == 1 and row['sktm'] == 1 and row['pendapatan'] < 1500000:
        return "DITERIMA"
    else:
        return "DITOLAK"

# Tambah kolom keputusan
df['status_keputusan'] = df.apply(hasil_keputusan, axis=1)

# ==== TAMPILKAN TABEL ====
print("=== TABEL HASIL SELEKSI MUSTAHIK ===")
display(df[['id','ktp','kk','sktm','pendapatan','status','status_keputusan']])

# ==== TOTAL STATUS ====
total_diterima = (df['status_keputusan'] == "DITERIMA").sum()
total_ditolak  = (df['status_keputusan'] == "DITOLAK").sum()

print(f"\nTotal DITERIMA : {total_diterima}")
print(f"Total DITOLAK  : {total_ditolak}")
```
